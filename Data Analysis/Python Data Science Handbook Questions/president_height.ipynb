```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv(
    "C:/Machine Learning Journey/Data Analysis/Python Data Science Handbook Questions/Files/president_heights.csv"
)
```

```python
data
```

```python
heights = np.array(data['height(cm)'])
print(heights)
```

Average height:

```python
print("Average height: ",heights.mean())
```

Median height:

```python
print("Median height: ",np.median(heights))
```

Standard deviation:

```python
print("Standard deviation: ",heights.std())
```

Min and max heights:

```python
print("Minimum height: ",heights.min())
print("Maximum height: ",heights.max())
```

Visualization:

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
plt.hist(heights)
plt.title("Heights distribution of US presidents")
plt.xlabel('height(cm)')
plt.ylabel('number')
```
