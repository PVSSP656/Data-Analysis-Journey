```python
import numpy as np
```

```python
arr = np.arange(0,10)
```

```python
arr
```

```python
print(arr+5)
```

```python
print(arr**arr)
```

Instead of showing fully errors, it's just showing small error/ a warning but still gives output for other elements in array

```python
print(arr/arr)
```

Sum of elements in an array

```python
print(arr.sum())
```

To get mean of an array

```python
print(arr.mean())
```

To get variance of elements in an array

```python
print(arr.var())
```

To get standard deviation of elements in an array

```python
print(arr.std())
```

All we did was for 1 dimensional array, now let's do for multiple and check

```python
arr2d = np.arange(0,25).reshape(5,5)
print(arr2d)
```

Let's check sum

```python
print(arr2d.sum())
```

Variance

```python
print(arr2d.var())
```

Standard deviation

```python
print(arr2d.std())
```

Now if we want to get sum of only columns/ rows:

```python
print(arr.sum(0)) # columns = 0
```

```python
print(arr.sum(1)) # rows = 1
```
