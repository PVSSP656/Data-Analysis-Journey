```python
import numpy as np
import pandas as pd
```

```python
myindex = ['USA','Canada','Mexico']
mydata = ['1776','1867','1821']
```

We see that the indices of elements in this array are normal integers

```python
mySeries = pd.Series(data = mydata)
mySeries
```

Now, we assign the list 'myindex' as index to mySeries - pd.Series automatically converts list to series

```python
mySeries = pd.Series(data = mydata, index = myindex)
mySeries
```

Above, we can see that the indices of mySeries are elements of myindex

To get elements on this series:

```python
mySeries[0]
```

In older versions of Pandas, this worked to get element at a particular position by numbers, now we've to use 'iloc'

```python
print(mySeries.iloc[0])
```

```python
mySeries['USA']
```

We can also create series from a dictionary

```python
ages = {'Sam':5, 'Frank':10, 'Spike':7}
pd.Series(ages)
```
