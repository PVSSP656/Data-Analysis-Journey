In new Pandas, this method got removed, instead use this:

If one_row is a DataFrame:
df = pd.concat([df, one_row], ignore_index=True)

if one_row is a Series:
df = pd.concat([df, one_row.to_frame().T], ignore_index=True)

If one_row is a dictionary:
df = pd.concat([df, pd.DataFrame([one_row])], ignore_index=True)
