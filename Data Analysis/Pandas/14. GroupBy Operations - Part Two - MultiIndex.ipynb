#### If we can use .loc, what's the use of cross-section? (.xs)

Suppose we want 4 cylinder cars, but for every single year, we can't use .loc, so it's 'cross-section' so we have to use '.xs'.

```python
year_cyl.xs(key=4,axis=0,level='cylinders')
```

```python
help(year_cyl.xs)
```

Suppose we want 6 and 8 of all sections together, since we can't pass a list in .xs, the best way to do is:

First filter out the 6 and 8 values from the original dataframe and then use groupby()

```python
df[df['cylinders'].isin([6,8])]
```

```python
df[df['cylinders'].isin([6,8])].groupby(['model_year','cylinders']).mean(numeric_only=True)
```

Let's see how to sort the multi-level index:

Now, it's sorted from earliest model_year to latest model_year, let's try reverse:

```python
year_cyl.sort_index(level='model_year',ascending=False)
```

### What we earlier did is specifying a same function for all the values, but in real life, it doesn't really makes sense, what we can do is, let's try giving separate aggregate functions to each and every column possible:

It's in old pandas:

```python
df.agg(['std', 'mean'])
```

In new Pandas:

```python
df.select_dtypes(include='number').agg(['std','mean'])
```

```python
df.select_dtypes(include='number').agg(['std','mean'])['mpg']
```

```python
df.select_dtypes(include='number').agg({'mpg':'std','cylinders':'mean'})
```

If multiple operations on a single column:

```python
df.select_dtypes(include='number').agg({'mpg':['max','mean']})
```

## Why does the first agg() result appear as a Series, but the second agg() result appear as a DataFrame? How does Pandas decide which type to return?

In first case, each column has one operation, so Pandas returns a Series.
And, a series is displayed as a simple list-like output.

Now, mpg has multiple operations (max and mean).
Pandas needs a table to store multiple results, so it returns a dataframe.

```python
df1 = df.select_dtypes(include='number').agg({'mpg': ['max','mean'],'weight': ['min','std']})
```

```python
df1.fillna('Not calculated').transpose()
```
