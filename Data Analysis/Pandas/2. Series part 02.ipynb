```python
import pandas as pd
import numpy as np
```

```python
# Imaginary Sales Data for 1st and 2nd Quarters for Global Company
q1 = {'Japan': 80, 'China': 450, 'India': 200, 'USA': 250}
q2 = {'Brazil': 100,'China': 500, 'India': 210,'USA': 260}
```

```python
sales_q1 = pd.Series(q1)
sales_q2 = pd.Series(q2)
```

```python
sales_q1
```

```python
sales_q2
```

```python
print(sales_q1['Japan'])
```

Now numerically, use iloc

```python
print(sales_q1.iloc[0])
```

To check what keys are available for us inside a series:

```python
sales_q1.keys()
```

We can perform broadcasted operations on them

```python
[1,2] * 2
```

We see that, in list, it got increased, but not doubled (the values), so in NumPy arrays:

```python
np.array([1,2]) * 2
```

Now we see that the operation got broadcasted so we got double of the array

So, in Pandas series, it's 'broadcasted operations' so:

```python
sales_q1 * 2
```

We see that the values got doubled, unlike lists - also any other operations

We will do sales_q1 + sales_q2 but notice that Japan is replaced by Brazil in sales_q2 so:

```python
sales_q1 + sales_q2
```

Now, to avoid NaN and fill up with some value, do this:

```python
sales_q1.add(sales_q2, fill_value= 0)
```

Now, we see that Brazil and Japan got its values
